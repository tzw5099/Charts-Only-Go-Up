```python
index_list = []
name_list = []
symbol_list = []

n = 488
while n<9898:
# for n,x in enumerate(df['symbol'][0:2000]):
    time.sleep(1)
    index = df['index'][n]
    symbol = df['symbol'][n]
    index_list.append(index)
    symbol_list.append(symbol)
    try:
        name = robin_stocks.stocks.get_name_by_symbol(symbol)
        name_list.append(name)
    except:
        name = ""
        name_list.append(name)
    print(n,index,symbol,name)
    n+=1
```

```python
robinhood_names_df = pd.DataFrame({'index_list': index_list,
'name_list': name_list,
'symbol_list':symbol_list})
```

```python
robinhood_names_df.to_csv("../reference_data/robinhood_names_df.csv")
```
