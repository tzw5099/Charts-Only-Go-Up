```python
def replace_name(x):
    if len(x)>40:
        x = x.replace(" limited "," ")\
        .replace(" of "," ")\
        .replace(", Inc."," ")\
        .replace(" by "," ")\
        .split("Depositary ", 1)[0].split("Series A", 1)[0].split("Series B", 1)[0].split("Series C", 1)[0].split("Series D", 1)[0].split("Perpetual Pref", 1)[0].split("Dep Sh Shares", 1)[0].split("1/", 1)[0].split("expiring", 1)[0].split("Callable ", 1)[0].split("Floating Rate", 1)[0].split("Non Cumulative", 1)[0].split("Cumulatiive", 1)[0].split("Limited", 1)[0].split("Depository", 1)[0].split("$", 1)[0].split("Cumulative", 1)[0].split("Depository", 1)[0].split("Series 2", 1)[0].split("Common Sh", 1)[0].split("represents", 1)[0].split("Perp", 1)[0].split("Preferred", 1)[0].replace(" ltd."," ")\
        .replace(" ltd"," ")\
        .replace(" Inc."," ")
        x = x.replace(" limited"," ")\
        .replace(", Inc."," ")\
        .replace(" Class V"," ")\
        .replace(" Class C"," ")\
        .replace(" Class A"," ").replace(" Covered"," ")
        x = x.replace(" Shares"," ")\
        .replace(", Holdings"," ")\
        .replace(" Corporation"," ")\
        .replace(" Corp"," ")
        x=x.replace(" Class B"," ")\
        .replace(", . ",", ").replace(" Bancorp,"," Bancorp").replace(" ome "," Income ").replace(", . ",", ").replace(", . ",", ").replace(", . ",", ")\
        .replace(" . "," ").replace(" Holdings."," ").replace("Business Machines","International Business Machines")\
        .replace(" Series A"," ")\
        .replace(" Series B"," ")\
        .replace(" Series C"," ")\
        .strip(' ').removesuffix(' .').removesuffix(',').removesuffix('-').removesuffix('(').removesuffix(' ').strip(' ')
        x=x.strip(' ').removesuffix(' .').removesuffix(',').removesuffix('-').removesuffix('(').removesuffix(' ').strip(' ')
    else:
            pass
    if x[-3:-1] == "  " or  x[-2:] == " .":
        x=x[:-2]
    x = x.strip(" ")
        
    return x
```
