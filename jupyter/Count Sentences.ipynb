```python
# https://stackoverflow.com/questions/3675318/how-to-replace-the-some-characters-from-the-end-of-a-string
def replace_last(source_string, replace_what, replace_with):
    head, _sep, tail = source_string.rpartition(replace_what)
    return head + replace_with + tail
```

```python
replace_last_list = ["SA", "Inc", "Corp.", " Inc.", "Group", "NV", "Ltd", "S.A.", "Limited", "SE", "LP", "Ltd.", "PLC","International,", "SCA"]
```

```python
def if_too_long(name,replace_what, replace_done, replacement_last=False):
#     print(name,replace_what, replace_done, replacement_last)
    if len(name) > 10:
        for x in replace_last_list:
            if replace_last == True:
                print("replace_last",replace_last)
                name = replace_last(name,x, replace_done)
            else:
                name = name.replace(x, replace_done)
    else:
        pass
#         replaced = name
    name = name.replace("Pharmaceuticals","Pharma").replace("Technologies","Tech").replace("Technology","Tech")
    
    way_too_long_list = ["Acquisition", "Holdings","Group",]
    way_too_long_last_list =  [" Associates", "International"]
    if len(name) > 30:
        for x in way_too_long_list:
            name = name.replace(x, replace_done)
        for x in way_too_long_last_list:
            name = replace_last(name,x, replace_done)
    else:
        pass
    return name
#     return replaced
```

```python
def url_safe(string):
    string=string.lower()
    for ch in [ '!','.',
                 '$',
                 '&',
                 "'",
                 '(',
                 ')',
                 '*',
                 '+',
                 ',',
                 '-',
                 '/',' ',' lp','inc.','acquisition','group','partners','inc.',' nv',' ltd', 'preferred','representing', 'shares','plc','holdings','corporation']:
       if ch in string:
          string=string.replace(ch," ")
    for ch in ['technologies',]:
       if ch in string:
          string=string.replace(ch,"tech")
    string = string.replace("  "," ").replace("pharmaceuticals","pharma").replace("international","int").replace("technologies","tech").replace("technology","tech").replace("--","-").replace("corp","").replace("incorporated","")
    split_strings = string.split(" ")
    
    max_len_string = 20
    if len("-".join(split_strings)) > max_len_string:
        split_strings = split_strings[0:5]
    if len("-".join(split_strings)) > max_len_string:
        split_strings = split_strings[0:4]
    if len("-".join(split_strings)) > max_len_string:
        split_strings = split_strings[0:3]
    if len("-".join(split_strings)) > max_len_string:
        split_strings = split_strings[0:2]
    if string[-1]=="-":
        string = string[:-1]
    final_string = "-".join(split_strings).replace("---","-").replace("--","-")
    if final_string[-1]=="-":
        final_string = final_string[:-1]
    if final_string[-4:]=="-inc":
        final_string = final_string[0:len(final_string)-len("-inc")]
    if final_string[-3:]=="-sa":
        final_string = final_string[0:len(final_string)-len("-sa")]
        
#     final_string = replace_last(final_string,"-", "")
    return final_string
```
